```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.tools import tool
```

```python
# tool create
@tool
def multiply(a: int, b: int) -> int:
    """ Given 2 numbers a and b this returns their product """
    return a * b
```

```python
print(multiply.invoke({"a": 3, "b": 4}))
```

```python
#tool binding

llm = ChatOpenAI()
tools = [multiply]
llm_with_tools = llm.bind(tools=tools)
```

Tool calling

```python
result = llm_with_tools.invoke("What is 3 times 4?")
```

```python
result.tool_calls
```

```python
input = result.tool_calls[0]['args']
```

Tool Execution

```python
multiply.invoke(input)
```

tool massage

```python
tool_message = multiply.invoke(result.tool_calls[0])
```

final full query content

```python
query = HumanMessage('can you multiply 3 with 1000')
```

```python
message = [query]
```

```python
message
```

```python
result = llm_with_tools.invoke(message)
```

```python
message.append(result)
```

```python
tool_message = multiply.invoke(result.tool_calls[0])
```

```python
message.append(tool_message)
```

```python
llm_with_tools.invoke(message).content
```
