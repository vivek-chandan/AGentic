# Custom Tools

```python
from langchain_core.tools import tool
```

```python
#step 1 - crearte a function
def multiply(a,b):
  """ multipy two numbers """ ##[important- Add Doc string] it help llm to understand what this tool does
  return a*b
```

```python
#step 2- add type hints
def multiply(a:int, b:int)->int:
  """multiply two numbers"""
  return a*b
```

```python
#step 3 - add tool decorator
@tool
def multiply(a:int, b:int)->int:
  """multiply two numbers"""
  return a*b
```

```python
results = multiply.invoke({"a":3, "b":4})
```

```python
print(results)
```

with help of structure Pydantic

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
  a:int = Field(required=True, description="the first number to add")
  b:int = Field(required=True, description="the second number to add")
```

```python
def multiply_func(a:int, b:int) ->int:
  return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
results = multiply_tool.invoke({'a':3,'b':4})
print(results)
```
